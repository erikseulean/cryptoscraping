```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from datetime import datetime as Date
from crypto.libs.marketcap import MarketCap
from crypto.libs.price import Price
from crypto.libs.coins import Coins
from crypto.libs.scraping.refresh import run as MarketCapInit
from crypto.libs import dateutils
```

```python
import numpy as np
import pandas as pd
```

```python
coins = Coins()
price = Price()
```

```python
today = Date.today()
```

```python
cryptopia_coins = coins.exchanges(["Cryptopia"])
```

```python
data = []
i = 0
for coin in cryptopia_coins:
    i = i + 1
    print ("{}/{}".format(i, len(cryptopia_coins)))
    if coin == "BTC": 
        continue
    ico_date = coins.coin_data(coin, "ICO")
    
    coin_age = coins.coin_data(coin, "AGE")
    px = price.current(coin, "USDT", exchange="")
    
    timeseries = price.historical(coin, "USD", ico_date, today)
    
    sorted_timeseries = timeseries.sort_index()
    sorted_timeseries = pd.DataFrame(sorted_timeseries)
    ico_price = sorted_timeseries.head(1)["close"].values[0]
    ico_date = pd.to_datetime(sorted_timeseries.head(1).index.values[0]).date()

    x_values = {
        2: {"found": False, "time": None, "date": None},
        3: {"found": False, "time": None, "date": None},
        4: {"found": False, "time": None, "date": None},
        5: {"found": False, "time": None, "date": None},
        7: {"found": False, "time": None, "date": None},
        10: {"found": False, "time": None, "date": None},
        15: {"found": False, "time": None, "date": None}}



    for date, close in sorted_timeseries.iterrows():
        for times, info in x_values.items():
            if close.values[0] >= ico_price * times and info["found"] == False:
                x_values[times]["date"] = pd.to_datetime(date).date()
                x_values[times]["time"] = (x_values[times]["date"] - ico_date).total_seconds() / (60*60*24)
                x_values[times]["found"] = True
    
    coin_return = timeseries.tail(1)["close"].values[0] / timeseries.head(1)["close"].values[0] - 1
    
    thirty_d_change_vs_usd = None
    thirty_d_change_vs_btc = None
    seven_d_change_vs_usd = None
    seven_d_change_vs_btc = None
    one_d_change_vs_usd = None
    one_d_change_vs_btc = None
    one_h_change_vs_usd = None
    one_h_change_vs_btc = None
    thirty_min_change_vs_usd = None
    thirty_min_change_vs_btc = None
    five_min_change_vs_usd = None
    five_min_change_vs_btc = None
    
    try:
        thirty_d_change_vs_usd = price.change(coin, "USD", bar_size = 30, bar_type = "d")
    except:
        pass
    
    try:
        thirty_d_change_vs_btc = price.change(coin, "BTC", bar_size = 30, bar_type = "d")
    except:
        pass
    
    try:
        seven_d_change_vs_usd = price.change(coin, "USD", bar_size = 7, bar_type = "d")
    except:
        pass
    
    try: 
        seven_d_change_vs_btc = price.change(coin, "BTC", bar_size = 7, bar_type = "d")
    except:
        pass
    
    try: 
        one_d_change_vs_usd = price.change(coin, "USD", bar_size = 1, bar_type = "d")
    except:
        pass
    
    try:
        one_d_change_vs_btc = price.change(coin, "BTC", bar_size = 1, bar_type = "d")
    except:
        pass
    
    try:
        one_h_change_vs_usd = price.change(coin, "USD", bar_size = 1, bar_type = "h")
    except:
        pass
    
    try:
        one_h_change_vs_btc = price.change(coin, "BTC", bar_size = 1, bar_type = "h")
    except:
        pass
    
    try:
        thirty_min_change_vs_usd = price.change(coin, "USD", bar_size = 30, bar_type = "m")
    except:
        pass
    
    try:
        thirty_min_change_vs_btc = price.change(coin, "BTC", bar_size = 30, bar_type = "m")
    except:
        pass
    
    try:
        five_min_change_vs_usd = price.change(coin, "USD", bar_size = 5, bar_type = "m")
    except:
        pass
    
    try:
        five_min_change_vs_btc = price.change(coin, "BTC", bar_size = 5, bar_type = "m")
    except:
        pass
    
    data.append({"Coin": coin, 
                 "Price": px, 
                 "ICO": ico_date, 
                 "ICO Price": ico_price, 
                 "Age": coin_age, 
                 "Total Return": coin_return, 
                 "2x": x_values[2]["time"],
                 "3x": x_values[3]["time"],
                 "4x": x_values[4]["time"],
                 "5x": x_values[5]["time"],
                 "7x": x_values[7]["time"],
                 "10x": x_values[10]["time"],
                 "15x": x_values[15]["time"],
                 "30d USD": thirty_d_change_vs_usd, 
                 "30d BTC": thirty_d_change_vs_btc,
                 "7d USD": seven_d_change_vs_usd, 
                 "7d BTC": seven_d_change_vs_btc,
                 "1d USD": one_d_change_vs_usd, 
                 "1d BTC": one_d_change_vs_btc,
                 "1h USD": one_h_change_vs_usd, 
                 "1h BTC": one_h_change_vs_btc,
                 "30m USD": thirty_min_change_vs_usd, 
                 "30m BTC": thirty_min_change_vs_btc,
                 "5m USD": five_min_change_vs_usd, 
                 "5m BTC": five_min_change_vs_btc })
    
    cols = [   "Coin", 
               "Price", 
               "ICO", 
               "ICO Price", 
               "Age", 
               "Total Return", 
               "2x",
               "3x",
               "4x",
               "5x",
               "7x",
               "10x",
               "15x",
               "30d USD", 
               "30d BTC",
               "7d USD", 
               "7d BTC",
               "1d USD", 
               "1d BTC",
               "1h USD", 
               "1h BTC",
               "30m USD", 
               "30m BTC",
               "5m USD", 
               "5m BTC"]
    
df = pd.DataFrame(data, columns=cols)
```
